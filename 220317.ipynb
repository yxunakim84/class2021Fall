- np.array()를 이용하여 Python에서 사용하는 Tuple(튜플)이나 List(리스트)를 입력으로 numpy.ndarray를 만들 수 있다.

* **np.arange() 와 np.linspace()를 이용하여 연속적인 데이터도 쉽게 생성할 수 있다.**
 * np.arange(): N 만큼 차이나는 숫자 생성
 * np.linspace(): N 등분한 숫자 생성

````python
x = np.arange(0,10,2) # 2는 increment 증가 정도 -> 0, 2, 4, 6, 8 (10 미만까지 이므로 10은 포함 x)
print(x)
print(type(x))
print(x.shape) #1d array임. 5개의 아이템이 들어있는 1d array
````

````python
x = np.linspace(0,10,6) # 0부터 10까지 같은 간격으로 총 6개의 숫자를 만들어라. linear하게 space를 줘서 만들라는 뜻
print(x)
print(type(x))
print(x.shape)
````

* **np.zeros(), np.ones(), np.empty()를 이용하여 다양한 차원의 데이터를 쉽게 생성할 수 있다.**

 * np.zeros(shape) : 0으로 구성된 N차원 배열 생성
 * np.ones(shape) : 1로 구성된 N차원 배열 생성
 * np.empty(shape) : 초기화되지 않은 N차원 배열 생성

````python
x = np.ones([2, 2, 3])
# 3차원 array를 1로 채워줌 (3d array)
print(x)
print(type(x))
print(x.shape)
````

* **Shape 변경 (Shape Manipulation)**
 * np.ndarray의 shape를 다양한 방법으로 변경할 수 있는데, .reshape()는 지정한 차원으로 변환을 할 수 있다. 데이터 원본은 변경시키지 않고 복사하여 연산한 결과가 return 된다.

````python
x = np.zeros([2, 2, 3])
print(x)
print(type(x))
print(x.shape)
x.reshape(3,2,2)
# [2, 2, 3] => [3, 2, 2] 로 변경
````

# **generate a sine curve**

hr.JPG

````python
plt.figure()

theta1 = np.linspace(0, 2*np.pi, 100)
# 0부터 2pi(360도)까지 100개의 숫자를 만들어라. 0~360도
# theta1에 있는 개수가 몇개인가? => 100개 
s1 = np.sin(theta1)
# s1에 있는 숫자의 개수는 몇개인가? => 100개
theta2 = np.linspace(0, 2*np.pi, 100) + np.pi/2
# 90(0+90)도~450(360+90)도
# theta2에 있는 개수가 몇개인가? => 100개
s2 = np.sin(theta2)

plt.subplot(2, 1, 1) # (rows, columns, panel number) (=> 칸을 만드는 것으로 생각하면 되는 것 같다.)
# 2행 1열로 만든 것 중, 첫번째 index 의 것을 선택
plt.plot(theta1, s1, '.')
# plot(x축, y축, 표기 방법)

plt.subplot(2, 1, 2) # (rows, columns, panel number) (=> 두번 째 칸을 만듦)
# 2행 1열로 만든 것 중, 두번째 index의 것을 선택
plt.plot(theta2, s2, '.')
````

````python
np.corrcoef(s1,s2)
````

```
corrcoef -> 상관관계 -1 <= r <= 1
r의 절댓값(|r|)이 1에 가까워질수록 상관관계가 높다.
r이 1 => 정비례 / r이 -1 => 반비례

[ s1과 s1의 상관정도, s1과 s2의 상관정도]
[ s2와 s1의 상관정도, s2와 s2의 상관정도]


자기 자신끼리(S,S or S2, S2) 는 위상차가 0도 이므로 cos 0 = 1 로 corrcoef가 당연히 1이다. S와 S2는 위상 차가 90도 이므로 cos 90 = 0 으로 corrcoef가 0이다.
(2.419... e-17(10의 -17승) = 거의 0에 가까움.)
```
